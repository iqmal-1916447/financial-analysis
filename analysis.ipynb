```python
URL_DATA = 'https://storage.data.gov.my/commodities/fuelprice.parquet'

df = pd.read_parquet(URL_DATA)
display(df)
```

```python
df['date'] = pd.to_datetime(df['date'])
display(df)
```

```python
# make date as index
df.set_index('date', inplace=True)
display(df)
```
