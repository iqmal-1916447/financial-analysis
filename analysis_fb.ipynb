```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ms_df = pd.read_csv('./microsoft.csv')
display(ms_df)
```

```python
# new column - price difference = Tomorrow price - Today price
ms_df['Price Diff'] = ms_df['Close'].shift(-1, axis=0) - ms_df['Close']
display( ms_df )
```

```python
ms_df.set_index('Date', inplace=True)
```

```python
ms_df.loc['2015-01-05', 'Price Diff']
```

```python
# new column - Daily Return = Today Price Diff / Today Close
ms_df['Daily Return'] = ms_df['Price Diff'] / ms_df['Close']
display(ms_df)
```

```python
ms_df.loc['2015-01-05', 'Daily Return']
```

```python
#Create a new column Direction. 
# if the price difference is larger than 0, denote as 1, otherwise, denote as 0,

ms_df['Direction'] = 0      # initial generalization
ms_df.loc[ ms_df['Price Diff'] > 0, 'Direction'] = 1
display( ms_df )
```

```python
ms_df.loc['2015-01-05'] [['Price Diff', 'Direction']]
```
